Cột có kiểu dữ liệu là int64: `Timestamp` 

Các cột có kiểu dữ liệu là float64: `Open`, `High`, `Low`, `Close`, `Volume_(BTC)`, `Volume_(Currency)`, `Weighted_Price` 

Cột `Timestamp` có kiểu dữ liệu chưa phù hợp: 

- Việc sử dụng Unix-time có thể dẫn đến khó hoặc không thể hiểu rõ ràng đối với những người sử dụng dữ liệu với mục đích tham khảo nhanh 

- Đối với dữ liệu time-series để có thể dễ dàng rút trích được ý nghĩa cũng như là để nhận xét được từ những  biểu đồ tạo ra bằng các phương pháp trực quan, quá trình trực quan cần dữ liệu thể hiện ngày tháng rõ ràng. 

$\Rightarrow$ Đề xuất xử lý và chuyển đổi thời gian từ Unix-time sang dạng datetime64 thông thường.

1. `Timestamp`: chuyển từ unix-time sang dateime 

2. Những dòng bị thiếu dữ liệu có các  cách xử lý: 

	+ Xóa những dòng này nhưng không đảm bảo trình tự logic của thời gian. 

	+ Điền vào những vị trí bị thiếu: 

	 	* Forward fill 

		* Backward fill 

$\Rightarrow$ Hướng giải quyết của nhóm: xóa những dòng bị thiếu

3. Sau khi xoá những dòng dữ liệu bị thiếu, thông tin về giá theo từng phút sẽ không đảm bảo tính liên tục

$\Rightarrow$ Gom nhóm theo ngày để đảm bảo tính liên tục của bộ dữ liệu


```python
# Import library
import pandas as pd
import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
df = pd.read_csv('./bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv')
df.head()
```

Cột `Timestamp` có kiểu dữ liệu lưu trữ là `int64`, chứa dữ liệu về thời gian kiểu Unix-time $\rightarrow$ Để giúp việc đọc và trực quan hoá dễ dàng hơn, cần chuyển cột này sang dạng `datetime`

```python
# Unix-time to datetime
df.Timestamp = pd.to_datetime(df.Timestamp, unit='s')
```

### Kiểm tra missing value

```python
# Missing value in dataset
df.isnull().sum()
```

```python
# Show all row with at least 1 missing value
nan_rows = df[df.isnull().any(axis=1)]
nan_rows
```

### Nhận xét:
+ Có **1243608** dòng đều bị thiếu dữ liệu ở các cột `['Open', 'High', 'Low', 'Close', 'Volume_(BTC)', 'Volume_(Currency)', 'Weighted_Price']`
+ Các dòng thiếu dữ liệu này thường là một vài phút hay vài giờ trong 1 ngày $\rightarrow$ Có thể xoá các dòng bị thiếu dữ liệu, nếu như không có ngày nào thiếu dữ liệu toàn bộ 24h

```python
nan_rows['day'] = nan_rows['Timestamp'].dt.date
# Count missing values by day
nan_rows_by_day = nan_rows.groupby('day')['Timestamp'].count().reset_index(name='count')

nan_rows_by_day.head()
```

```python
# Find day have 1440 missing values (1 day = 1440 min)
nan_rows_by_day[nan_rows_by_day['count'] == 1440].head()
```

### Nhận xét:
+ Không có bất kỳ ngày nào bị thiếu toàn bộ dữ liệu $\rightarrow$ Có thể xoá tất cả các dòng thiếu dữ liệu

```python
# Drop NaN
df.dropna(axis=0, how='any', inplace=True)
```

### Tạo bảng gom nhóm dataset theo ngày

```python
df_day = df.copy()
df_day['Timestamp'] = df_day['Timestamp'].dt.tz_localize(None)
df_day = df_day.groupby([pd.Grouper(key='Timestamp', freq='D')]).agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume_(BTC)': 'sum',
    'Volume_(Currency)': 'sum',
    'Weighted_Price': 'mean'
}).reset_index()
df_day = df_day.set_index('Timestamp')
df_day.head()
```

### Xuất dataset theo ngày ra file "BTC-Historical-Price.csv"

```python
df_day.to_csv('BTC-Historical-Price.csv')
```
